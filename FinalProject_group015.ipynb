## Background & Prior Work

Depression is a common, but serious mood disorder in the U.S., especially among adolescents and young adults. # Dataset(s)

- Dataset Name: Suicide Mortality by State
- Link to the dataset: https://www.cdc.gov/nchs/pressroom/sosmap/suicide-mortality/suicide.htm
- Number of observations: 400 (50 states for 8 years)

This dataset from the CDC contains suicide mortality rates (in # of deaths per 100,000), and absolute deaths by suicide, for each state, for select years.

- Dataset Name: Average Incidence of Depression by State, 2014-2018
- Link to the dataset: https://quotewizard.com/news/posts/states-with-highest-depression-rates
- Number of observations: 50 states

This dataset averages the incidence of clinical depression (by percentage) among the general population between the years 2014 and 2018, by state. It was prepared by QuoteWizard for an article about mental health and health insurance, and was calculated using aggregated data obtained from the Center for Disease Control and Prevention (CDC).

- Dataset Name: Average Annual Sunshine by State
- Link to the dataset: https://www.currentresults.com/Weather/US/average-annual-state-sunshine.php?fbclid=IwAR348oIxJc5iUcAzeF9fYuftndUuTdPQ7QHbchC4uXaYgr1snCrVDZjBdOs
- Number of observations: 50 states

Sunshine in not something that most weather stations report. This dataset therefore uses available data from large cities within each state to represent the average number of clear days (days that have <30% cloudcover) per year, total hours of sunshine per year, and percent sunshine (percent of time between dawn and disk the sun hits the ground) per yer, for that state. This dataset uses historic data from the National Climate Data Center and the World Data Center for Meteorology in computing its averages.    

- Dataset Name: World Bank Group, Climate Change Knowledge Portal Annual Climate Data for US, Historical Reference Period 1901-2020
- Link to the dataset: https://climateknowledgeportal.worldbank.org/download-data
- Number of observations: 120 years (data for 50 states per year)

This dataset was obtained from the World Bank Group's Climate Change Knowledge Portal, and contains, for each state in the US, the minimum tempature (in &deg;C), maximum tempature (in &deg;C), average tempature (in &deg;C), and total precipitation (in mm) for the years 1901-2020, based on data obtained by The Climate Research Unit (CRU). This dataset is split between four files: each variable is contained in a different file.  


We will combine the above datasets in the following way: from the 'Suicide Mortality by State' dataset, we will average the five columns corresponding to the years 2014-2018, and include this as a column in our final dataset; since the 'Average Incidence of Depression by State, 2014-2018' already is averaged for the years 2014-2018, we will take this column, adjust the given percentage to incidence per 100,000 (for comparison with 'Suicide Mortality by State' dataset), and include this as a column in our final dataset; from the 'Average Annual Sunshine by State' dataset, we are interested only in average yearly rainfall and average total hours of yearly sunshine, so we will include only these columns in our final dataset; from the 'Climate Change Knowledge Portal Annual Climate Data for US', we will average the columns corresponding to years 2014-2018 for min. temp., max. temp., average temp., and rainfall, and include this data for each state as seperate columns in our final dataset. Each dataset that will be combined to form our final dataset is indexed by state, making the correspondance between them intuative and easy.

```python
#imports

import numpy as np
import pandas as pd
from tabulate import tabulate

import patsy
import statsmodels.api as sm

import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (20, 7) #increase figure size

import seaborn as sns
sns.set(style='white', font_scale=2) #set style

from scipy.stats import ttest_ind

# read in raw datasets from csv files
dataset_dir       = 'datasets/'
depression_df_raw = pd.read_csv(dataset_dir + 'Depression_2014-2018.csv', usecols = ['State','Depression Rate'], index_col = 0)
suicide_df_raw    = pd.read_csv(dataset_dir + 'Suicide_2005-2020.csv', usecols = ['YEAR','STATE','RATE'])
max_temp_df_raw   = pd.read_csv(dataset_dir + 'Temp_Max_1901-2020.csv', skiprows = 1, index_col=0).T
min_temp_df_raw   = pd.read_csv(dataset_dir + 'Temp_Min_1901-2020.csv', skiprows = 1, index_col=0).T
avg_temp_df_raw   = pd.read_csv(dataset_dir + 'Temp_Mean_1901-2020.csv', skiprows = 1, index_col=0).T
sun_hours_df_raw  = pd.read_csv(dataset_dir + 'Sunshine.csv', index_col=0)
rain_df_raw       = pd.read_csv(dataset_dir + 'Rainfall_1901-2020.csv', skiprows = 1, index_col=0).T
```

## Basic Data Descriptions

```python
# name of data categories we want to display
data_names = ['shape', 'data types', 'contains NaN']

# list of lists - each element is a list describing a dataset with the following format:
# [0 = dataset shape, 1 = dataset data types, 2 = dataset contains NaN]
basic_data = [
    [str(depression_df_raw.shape), str(depression_df_raw.dtypes), str(depression_df_raw.isnull().any())],
    [str(suicide_df_raw.shape), str(suicide_df_raw.dtypes), str(suicide_df_raw.isnull().any())],
    [str(max_temp_df_raw.shape), str(max_temp_df_raw.dtypes), str(max_temp_df_raw.isnull().any())],
    [str(min_temp_df_raw.shape), str(min_temp_df_raw.dtypes), str(min_temp_df_raw.isnull().any())],
    [str(avg_temp_df_raw.shape), str(avg_temp_df_raw.dtypes), str(avg_temp_df_raw.isnull().any())],
    [str(sun_hours_df_raw.shape), str(sun_hours_df_raw.dtypes), str(sun_hours_df_raw.isnull().any())],
    [str(rain_df_raw.shape), str(rain_df_raw.dtypes), str(rain_df_raw.isnull().any())]
]

# name of each dataset being described
dataset_names = ['depression_df_raw', 'suicide_df_raw', 'max_temp_df_raw', 'min_temp_df_raw', 'avg_temp_df_raw', 'sun_hours_df_raw', 'rain_df_raw']

# create table to print
print(tabulate(basic_data, headers=data_names, tablefmt='fancy_grid', showindex=dataset_names))
```

Clean and organize temperature and rainfall data. The range was filtered to be from 2014-2018 and was done through removing superfluous columns. Extra rows were also removed, specifically 'United States' and 'District of Columbia' since our research question pertains to the state level and not anything greater or lesser. Also, a new column was added to each dataframe containing mean across the years.

```python
# restrict range to 2014-2018 by removing superfluous columns
years_to_keep   = [2014, 2015, 2016, 2017, 2018]
max_temp_df_res  =  max_temp_df_raw.loc[:, max_temp_df_raw.columns.intersection(years_to_keep)].copy()
min_temp_df_res =  min_temp_df_raw.loc[:, min_temp_df_raw.columns.intersection(years_to_keep)].copy()
avg_temp_df_res =  avg_temp_df_raw.loc[:, avg_temp_df_raw.columns.intersection(years_to_keep)].copy()
rain_df_res     =  rain_df_raw.loc[:, rain_df_raw.columns.intersection(years_to_keep)].copy()

# remove extra rows
rows_to_remove = ['United States', 'District of Columbia']
max_temp_df_res.drop(rows_to_remove, inplace = True)
min_temp_df_res.drop(rows_to_remove, inplace = True)
avg_temp_df_res.drop(rows_to_remove, inplace = True)
rain_df_res.drop(rows_to_remove, inplace = True)

# add new column to each dataframe containing mean across years
max_temp_df_res['max_temp'] = max_temp_df_res[years_to_keep].mean(axis=1)
min_temp_df_res['min_temp'] = min_temp_df_res[years_to_keep].mean(axis=1)
avg_temp_df_res['avg_temp'] = avg_temp_df_res[years_to_keep].mean(axis=1)
rain_df_res['rain']         = rain_df_res[years_to_keep].mean(axis=1)
```

Clean and organize sunshine data. Any '-' values were replaced with '-1', column titled 'Total Hours' and 'Clear Days' were renamed 'sun_hours' and 'clear_days.' Lastly, the 'sun_hours' type was converted to integer.

```python
sun_hours_df_raw.replace('–', '-1', inplace = True)
sun_hours_df_res = sun_hours_df_raw[['Total Hours','Clear Days']].copy()
sun_hours_df_res.rename(columns={'Total Hours':'sun_hours', 'Clear Days':'clear_days'}, inplace=True)
sun_hours_df_res['sun_hours'] = sun_hours_df_res['sun_hours'].astype('int')
```

Clean and organize depression data. The depression percentage was converted to number of cases per 100,000 in order to have a better comparison with suicide. The type of the 'Depression Rate' was also set to integer.

```python
#convert depression percentage to # of cases per 100,000 (for comparison with suicide)
depression_df_raw['Depression Rate'] = depression_df_raw['Depression Rate'].str.rstrip('%').astype('float')*1000
depression_df_raw['Depression Rate'] = depression_df_raw['Depression Rate'].astype('int')
```

Clean and organize suicide data. Data was filtered so range was between 2014 and 2018 by removing superfluous rows. State abbreviations were converted to state names and the suicide average for each state was calculated as dictionary, then convert to dataframe.

```python
# restrict range to 2014-2018 by removing superfluous rows
years_to_keep = [2014, 2015, 2016, 2017, 2018]
suicide_df_raw.drop(suicide_df_raw[~suicide_df_raw['YEAR'].isin(years_to_keep)].index, inplace = True)

# State abbreviations need to be converted to state names. 
# State name-abbreviation dictionary from https://gist.github.com/rogerallen/1583593
us_state_to_abbrev = {
    "Alabama": "AL",
    "Alaska": "AK",
    "Arizona": "AZ",
    "Arkansas": "AR",
    "California": "CA",
    "Colorado": "CO",
    "Connecticut": "CT",
    "Delaware": "DE",
    "Florida": "FL",
    "Georgia": "GA",
    "Hawaii": "HI",
    "Idaho": "ID",
    "Illinois": "IL",
    "Indiana": "IN",
    "Iowa": "IA",
    "Kansas": "KS",
    "Kentucky": "KY",
    "Louisiana": "LA",
    "Maine": "ME",
    "Maryland": "MD",
    "Massachusetts": "MA",
    "Michigan": "MI",
    "Minnesota": "MN",
    "Mississippi": "MS",
    "Missouri": "MO",
    "Montana": "MT",
    "Nebraska": "NE",
    "Nevada": "NV",
    "New Hampshire": "NH",
    "New Jersey": "NJ",
    "New Mexico": "NM",
    "New York": "NY",
    "North Carolina": "NC",
    "North Dakota": "ND",
    "Ohio": "OH",
    "Oklahoma": "OK",
    "Oregon": "OR",
    "Pennsylvania": "PA",
    "Rhode Island": "RI",
    "South Carolina": "SC",
    "South Dakota": "SD",
    "Tennessee": "TN",
    "Texas": "TX",
    "Utah": "UT",
    "Vermont": "VT",
    "Virginia": "VA",
    "Washington": "WA",
    "West Virginia": "WV",
    "Wisconsin": "WI",
    "Wyoming": "WY",
    "District of Columbia": "DC",
    "American Samoa": "AS",
    "Guam": "GU",
    "Northern Mariana Islands": "MP",
    "Puerto Rico": "PR",
    "United States Minor Outlying Islands": "UM",
    "U.S. Virgin Islands": "VI",
}
abbrev_to_us_state = dict(map(reversed, us_state_to_abbrev.items()))
suicide_df_raw['STATE'] = suicide_df_raw['STATE'].replace(abbrev_to_us_state)

#calculate average for each state as dictionary, then convert to df
suicide_avg = dict([(state, suicide_df_raw[suicide_df_raw['STATE'] == state]['RATE'].mean()) for state in suicide_df_raw['STATE'].unique()])
suicide_df_res = pd.DataFrame.from_dict(suicide_avg, orient='index') 
suicide_df_res.rename(columns={0:'suicide'}, inplace=True)
```

Create final, aggregated dataframe. After processing all the data, a dataframe was initialized and then everything was aggregated into this final dataframe

```python
#initialize final dataframe df
cols = ['max_temp', 'min_temp', 'avg_temp', 'rain', 'sun_hours']
rows = max_temp_df_res.index.copy()
df = pd.DataFrame(index=rows)

#put together final dataframe df
df = df.join(max_temp_df_res['max_temp'].copy(), how = 'left')
df = df.join(min_temp_df_res['min_temp'].copy(), how = 'left')
df = df.join(avg_temp_df_res['avg_temp'].copy(), how = 'left')
df = df.join(rain_df_res['rain'].copy(), how = 'left')
df = df.join(sun_hours_df_res.copy(), how = 'left')
df = df.join(depression_df_raw, how = 'left')
df.rename(columns={'Depression Rate':'depression'}, inplace=True)
df = df.join(suicide_df_res, how = 'left')
```

## Linear Regression

### Linearity - Average Temperature vs Depression

```python
# Average Temperature vs Depression
outcome, predictors = patsy.dmatrices('depression ~ avg_temp', df)
mod = sm.OLS(outcome, predictors)
res = mod.fit() 
print(res.summary(slim=True))
```

1.4% (r-squared=0.014) of the variation in Depression is explained by Average Temperature, which is very weak. Therefore, there is virtually no correlation between average temperature and depression. The p-value is 0.407, which means that it is not statistically significant. Therefore, we fail to reject the null hypothesis. This means there is an insignificant relationship between Average Temperature and Depression. 

### Linearity - Average Temperature vs Suicides

```python
# Average Temperature vs Suicides
outcome, predictors = patsy.dmatrices('suicide ~ avg_temp', df)
mod = sm.OLS(outcome, predictors)
res = mod.fit() 
print(res.summary(slim=True))
```

12.3% (r-squared=0.123) of the variation in Suicides is explained by Average Temperature, which is a small r-squared value. ## Classification
In this section we will take a different approach by using classification and running some t-tests to see if we can see any relationships between **rainfall, sun hours, and average temperature** and **suicide and rainfall**.

### Depression and Suicide in Rainy vs Not Rainy States
Here we will split the dataset into two parts - "rainy" states and "dry" states. The threshold for being considered a rainy state will be the median **rainfall** value in the entire dataset. Any state experiencing more annual rainfall on average than this threshold will be considered a "rainy" state while the rest will be considered "dry" states. Then we will use a t-test to see if there are any differences in the depression and suicide rates between the two types of states.

In order to run a t-test we assume:
1. continuous data
2. normally distributed
3. equal variance between 2 groups
4. data across each group not paired

**Null Hypothesis** = no difference in depression/suicide rates in rainy/dry states  
**Alternative Hypothesis** = there is a difference in depression/suicide rates in rainy/dry states

```python
# classify each state as either rainy or not-rainy
median_rainfall = df['rain'].median()
rain_df = df[['rain', 'suicide', 'depression']].copy()
rain_df['rainy_state'] = ['rainy' if x >= median_rainfall else 'dry' for x in rain_df['rain']]
```

```python
# ensure suicide data is normally distributed
sns.displot(rain_df, x='suicide', hue='rainy_state', kind='kde')
```

```python
# get suicide data for rainy and not-rainy states
suicide_rainy = rain_df[rain_df['rainy_state'].isin(['rainy'])]['suicide']
suicide_dry = rain_df[rain_df['rainy_state'].isin(['dry'])]['suicide']
```

```python
# ensure variance of suicide data is close to equal - as a rule of thumb ratio should be less than 4 for t-test
suicide_rainy_variance = suicide_rainy.var()
suicide_dry_variance = suicide_dry.var()
print('Ratio of Suicide Variances Between Rainy/Dry States =', suicide_dry_variance / suicide_rainy_variance)
```

```python
# run t-test for suicide data
ttest_ind(suicide_rainy, suicide_dry)
```

After running our t-test, we see that our p-value is very low. Therefore we **reject the Null Hypothesis** and conclude that **there is indeed a difference** between the suicide rate of rainy and dry states based on this sample.

```python
# ensure depression data is normally distributed
sns.displot(rain_df, x='depression', hue='rainy_state', kind='kde')
```

```python
# get depression data for rainy and not-rainy states
depression_rainy = rain_df[rain_df['rainy_state'].isin(['rainy'])]['depression']
depression_dry = rain_df[rain_df['rainy_state'].isin(['dry'])]['depression']
```

```python
# ensure variance of depression data is close to equal - as a rule of thumb ratio should be less than 4 for t-test
depression_rainy_variance = depression_rainy.var()
depression_dry_variance = depression_dry.var()
print('Ratio of Depression Variances Between Rainy/Not-Rainy States =', depression_rainy_variance / depression_dry_variance)
```

```python
# run t-test for suicide data
ttest_ind(depression_rainy, depression_dry)
```

Based on the high p-value from the t-test on this sample data, we **cannot reject the Null Hypothesis** and must assume that there is **no difference** in depression rates between rainy and dry states.
